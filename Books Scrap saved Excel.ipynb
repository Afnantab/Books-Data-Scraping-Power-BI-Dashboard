```python
import requests
from bs4 import BeautifulSoup
import openpyxl

excel = openpyxl.Workbook()
sheet = excel.active
sheet.title = "Books Scraping"
sheet.append(["Title", "Price", "Availability", "Rating", "Category", "Product Page URL"])

BASE_URL = "http://books.toscrape.com/catalogue/page-{}.html"

for page_num in range(1, 51):
    print(f"Scraping page {page_num}...")
    url = BASE_URL.format(page_num)
    page = requests.get(url).text
    soup = BeautifulSoup(page, "html.parser")

    
    books = soup.select("h3 a")

    for book in books:
        book_link = book["href"].replace("../../../", "")
        book_url = "http://books.toscrape.com/catalogue/" + book_link

        
        book_page = requests.get(book_url).text
        book_soup = BeautifulSoup(book_page, "html.parser")

        title = book_soup.find("div", class_="product_main").h1.text.strip()
        price = book_soup.find("p", class_="price_color").text.strip()
        availability = book_soup.find("p", class_="instock availability").get_text(strip=True)
        rating_tag = book_soup.find("p", class_="star-rating")
        rating = rating_tag["class"][1] if rating_tag else "Not Rated"

        
        category = book_soup.select("ul.breadcrumb li a")[-1].text.strip()

        
        sheet.append([title, price, availability, rating, category, book_url])
        print(f"Saved: {title}")


excel.save("books_data.xlsx")
```
